```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]="1"
```

```python
import cv2
import numpy as np
from mtcnn import MTCNN
```

```python
detector = MTCNN()
```

```python
def get_faces(frame):
    
    results = detector.detect_faces(frame)
    faces = []

    for i in range(len(results)):
      try:
        if results[i]['confidence'] > 0.95:

          x1, y1, w, h = results[i]['box']
          x2, y2 = x1 + w, y1 + h

          faces.append({
              'x1': x1,
              'y1': y1,
              'x2': x2,
              'y2': y2,
              'confidence': results[i]['confidence']
          })

      except:
        continue

    return faces
```

```python
cap = cv2.VideoCapture(0)

w, h = 160, 96

while cap.isOpened():
    
    ret, frame = cap.read()
    
    if not ret:
        break
    
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame_resize = cv2.resize(frame_rgb, (w, h))
    results = get_faces(frame_resize)
        
    for face in results:
        x1 = int((face['x1']/w) * frame.shape[1])
        x2 = int((face['x2']/w) * frame.shape[1])
        y1 = int((face['y1']/h) * frame.shape[0])
        y2 = int((face['y2']/h) * frame.shape[0])
    
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
        
    cv2.imshow('frame', frame)
    
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
