```python
import cv2
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
facenet_model = keras.models.load_model("facenet_keras.h5")
```

```python
def read_image(path):
    image = cv2.imread(path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_resize = cv2.resize(image_rgb, (160, 160))
    return image_resize
```

```python
def get_embeddings(face, facenet_model):

    face = cv2.resize(face, (160, 160))/255    
    face = face[None, ...]    
    
    return facenet_model.predict(face)[0]
```

```python
def reduction(x):
    pca.fit(x)
    return pca.fit_transform(x)
```

```python
eu = read_image('eu.jpeg')
outroeu = read_image('outroeu.jpg')
ufpi = read_image('ufpi.jpg')
```

```python
f, axarr = plt.subplots(3)
axarr[0].imshow(eu)
axarr[1].imshow(outroeu)
axarr[2].imshow(ufpi)
```

```python
cap = cv2.VideoCapture(0)

w, h = 160, 96
pca = PCA(n_components=2)
photo = read_image()
emb_photo = get_embeddings(photo, facenet_model)

fig = plt.figure()

while cap.isOpened():
    
    ret, frame = cap.read()
    
    if not ret:
        break
    
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame_resize = cv2.resize(frame_rgb, (w, h))
    results = get_faces(frame_resize)
        
    for face in results:
        x1 = int((face['x1']/w) * frame.shape[1])
        x2 = int((face['x2']/w) * frame.shape[1])
        y1 = int((face['y1']/h) * frame.shape[0])
        y2 = int((face['y2']/h) * frame.shape[0])
        
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
        face = frame[y1:y2, x1:x2]
        emb = get_embeddings(face, facenet_model)
        
        x = np.array([emb, emb_photo])
        new_x = reduction(x)
        plot(new_x[0][0], new_x[0][1], 'bo')
        plot(new_x[1][0], new_x[1][1], 'ro')
        
    cv2.imshow('frame', frame)
    
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
